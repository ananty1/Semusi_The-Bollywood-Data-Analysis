## Preprocess the text

```python
# Read the script from a file
with open('Pink_file.txt', 'r', encoding='utf-8') as file:
    script_text = file.read()

# # Split the script into scenes (assuming scenes are separated by a line break)
# scenes = script_text.split('\n')  # Adjust the delimiter based on your script format


# # Print the movie plots

```

### Remove Special Characters and Extra Spaces

```python
import re


# Remove special characters and extra spaces
cleaned_script = re.sub(r'\W+', ' ', script_text)
cleaned_script = re.sub(r'\s+', ' ', cleaned_script)
```

### Tokenization: Tokenize the text into words or sentences for further analysis.

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize

# Tokenize into words
words = word_tokenize(cleaned_script)

# Tokenize into sentences
sentences = sent_tokenize(cleaned_script)
```

Remove Stopwords: Remove common stopwords (like 'the', 'and', 'is') that may not contribute much to the analysis.

```python
from nltk.corpus import stopwords

# Download stopwords (run this once)
import nltk
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
```

### Lemmatization or Stemming: Reduce words to their base or root form using lemmatization or stemming.

python

```python
from nltk.stem import WordNetLemmatizer, PorterStemmer
import nltk
nltk.download('wordnet')

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```python
normalized_words = [word.lower() for word in stemmed_words]
```

## Using a pretrained distilled bird sentence based stereotype creater

```python
import pandas as pd
dataset = pd.read_csv("/content/coref_plot.csv",index_col=0).iloc[:1,:]
```

```python
from tqdm.notebook import tqdm

# Function to determine stereotype words
def StereoWords(sentence):
    Dict = {'unrelated': [], 'stereotype_gender': [], 'anti-stereotype_gender': [],
            'stereotype_race': [], 'anti-stereotype_race': [], 'stereotype_profession': [],
            'anti-stereotype_profession': [], 'stereotype_religion': [], 'anti-stereotype_religion': []}
    # You need to define nlp and import the necessary libraries for this part
    for word in tqdm(sentence.split(), desc="Processing Words", leave=False):
        output = nlp(word)
        Dict[output[0]['label']].append(word)

    if 'stereotype' in nlp(sentence)[0]['label']:
        return 'stereotype', Dict
    else:
        return 'non_stereotype', Dict

# Apply the function to each row
tqdm.pandas(desc="Processing Rows")
type, dicti = zip(*dataset['Coref Plot'].progress_apply(StereoWords))
dataset['type'], dataset['determining_words'] = type, dicti

# Create new columns for each stereotype type
dataset['unrelated'] = [words['unrelated'] for words in dataset['determining_words']]
dataset['stereotype_gender'] = [words['stereotype_gender'] for words in dataset['determining_words']]
dataset['anti-stereotype_gender'] = [words['anti-stereotype_gender'] for words in dataset['determining_words']]
dataset['stereotype_race'] = [words['stereotype_race'] for words in dataset['determining_words']]
dataset['anti-stereotype_race'] = [words['anti-stereotype_race'] for words in dataset['determining_words']]
dataset['stereotype_profession'] = [words['stereotype_profession'] for words in dataset['determining_words']]
dataset['anti-stereotype_profession'] = [words['anti-stereotype_profession'] for words in dataset['determining_words']]
dataset['stereotype_religion'] = [words['stereotype_religion'] for words in dataset['determining_words']]
dataset['anti-stereotype_religion'] = [words['anti-stereotype_religion'] for words in dataset['determining_words']]

# Print the updated dataframe
print(dataset)


```
