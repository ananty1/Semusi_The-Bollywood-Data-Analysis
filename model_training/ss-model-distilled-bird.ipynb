```python
from transformers import pipeline

nlp = pipeline("text-classification", model="wu981526092/Sentence-Level-Stereotype-Detector", tokenizer="wu981526092/Sentence-Level-Stereotype-Detector")
# nlp = pipeline("text-classification")
from tqdm.notebook import tqdm

# Function to determine stereotype words
def StereoWords(sentence):
    Dict = {'unrelated': [], 'stereotype_gender': [], 'anti-stereotype_gender': [],
            'stereotype_race': [], 'anti-stereotype_race': [], 'stereotype_profession': [],
            'anti-stereotype_profession': [], 'stereotype_religion': [], 'anti-stereotype_religion': []}
    # You need to define nlp and import the necessary libraries for this part
    for word in tqdm(sentence.split(), desc="Processing Words", leave=False):
        output = nlp(word)
        Dict[output[0]['label']].append(word)
        
    if 'stereotype' in nlp(sentence)[0]['label']:
        return 'stereotype', Dict
    else:
        return 'non_stereotype', Dict

# Apply the function to each row
tqdm.pandas(desc="Processing Rows")
type, dicti = zip(*dataset['Coref Plot'].progress_apply(StereoWords))
dataset['type'], dataset['determining_words'] = type, dicti

# Create new columns for each stereotype type
dataset['unrelated'] = [words['unrelated'] for words in dataset['determining_words']]
dataset['stereotype_gender'] = [words['stereotype_gender'] for words in dataset['determining_words']]
dataset['anti-stereotype_gender'] = [words['anti-stereotype_gender'] for words in dataset['determining_words']]
dataset['stereotype_race'] = [words['stereotype_race'] for words in dataset['determining_words']]
dataset['anti-stereotype_race'] = [words['anti-stereotype_race'] for words in dataset['determining_words']]
dataset['stereotype_profession'] = [words['stereotype_profession'] for words in dataset['determining_words']]
dataset['anti-stereotype_profession'] = [words['anti-stereotype_profession'] for words in dataset['determining_words']]
dataset['stereotype_religion'] = [words['stereotype_religion'] for words in dataset['determining_words']]
dataset['anti-stereotype_religion'] = [words['anti-stereotype_religion'] for words in dataset['determining_words']]


```
